```python
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
from datetime import *

now = datetime.now()


while True:

    #draw clock body
    image = np.zeros((500, 500, 3), np.uint8) 
    cv2.circle(image,(250,250),150,(0,0,225),3)
    
    #Put daye and time
    now = datetime.now()
    time_now = now.strftime("%H:%M:%S")
    today = date.today()
    time="now =", time_now
    cv2.cv2.putText(image, f"today : {today}", (5,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)
    cv2.cv2.putText(image, f"now :{time_now}",  (5,70), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,0), 2)

    #draw hours' signs
    for r in range(0,360,30):
        x=int(250+150*math.cos(math.radians(r)))
        y=int(250+150*math.sin(math.radians(r)))
        cv2.circle(image,(x,y),4,(0,225,255),2)

        
    #draw minutes' signs
    for r in range(0,360,6):
        x0=int(250+150*math.cos(math.radians(r)))
        y0=int(250+150*math.sin(math.radians(r)))
        x1=int(250+147*math.cos(math.radians(r)))
        y1=int(250+147*math.sin(math.radians(r)))
        
        cv2.line(image,(x0,y0),(x1,y1),(255,255,255),1)
        
     #put hour numbers
    h=3
    for r in range(0,360,30):
        x=int(250+150*math.cos(math.radians(r)))
        y=int(250+150*math.sin(math.radians(r)))
        if h in [3]:
            cv2.cv2.putText(image, str(h), (x+6,y+10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [4,5]:
            cv2.cv2.putText(image, str(h), (x+6,y+25), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [6]:
            cv2.cv2.putText(image, str(h), (x-10,y+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [7,8]:
            cv2.cv2.putText(image, str(h), (x-30,y+25), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [9]:
            cv2.cv2.putText(image, str(h), (x-30,y+10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [10,11]:
            cv2.cv2.putText(image, str(h), (x-50,y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [12]:
            cv2.cv2.putText(image, str(h), (x-20,y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        elif h in [1,2]:
            cv2.cv2.putText(image, str(h), (x+6,y-5), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
        h+=1
        if h>12:
            h=1
            
            
#------------------------------------------------------------------------------

    #draw clockwise
    cv2.circle(image,(250,250),10,(50,50,50),-1)
    
    #seconds
    s=int(now.second)
    second_angle = math.fmod(s * 6 + 270, 360)
    x_s=int(250+140*math.cos(math.radians(second_angle)))
    y_s=int(250+140*math.sin(math.radians(second_angle)))
    cv2.line(image,(250,250),(x_s,y_s),(120, 120, 120), 1)
    
    
    
    #minutes
    m=now.minute
    minute_angle = math.fmod((m * 6)+(s/12)+ 270, 360)
    x_m=int(250+125*math.cos(math.radians(minute_angle)))
    y_m=int(250+125*math.sin(math.radians(minute_angle)))
    cv2.line(image,(250,250),(x_m,y_m),(150, 255, 150), 2)
    
    #hours
    h=math.fmod(now.hour, 12)
    hour_angle = math.fmod((h*30) + (m/2) + 270, 360)
    x_h=int(250+115*math.cos(math.radians(hour_angle)))
    y_h=int(250+115*math.sin(math.radians(hour_angle)))
    cv2.line(image,(250,250),(x_h,y_h),(220, 120, 50), 2)
            
    cv2.imshow("Clock",image)    
    k=cv2.waitKey(1000)
    if k==27:
        break
cv2.destroyAllWindows()
```
